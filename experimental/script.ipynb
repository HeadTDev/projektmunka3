### Hang kinyerése a videóból

### Szöveg kinyerése a hangfájlból

```python
model = whisper.load_model("large-v3-turbo")
result = model.transcribe("./extracted_audio/2.wav")
print(result["text"])
with open("./extracted_lyrics/lyrics.txt", "w", encoding="utf-8") as file:
    file.write(result["text"])
```

### Szöveg angolról magyarra fordítása

```python
load_dotenv()
api_key = os.getenv("DEEPL_API_KEY")

translator = deepl.Translator(api_key)

def translate_to_hungarian(english_text):
    result = translator.translate_text(english_text, target_lang="HU")
    return result.text

english_text = open("lyrics.txt", "r", encoding="utf-8").read()
hungarian_text = translate_to_hungarian(english_text)
print(hungarian_text)
with open("./translated_lyrics/lyrics.txt", "w", encoding="utf-8") as file:
    file.write(hungarian_text)
```

```python
load_dotenv()
api_key = os.getenv("DEEPL_API_KEY")

translator = deepl.Translator(api_key)

def translate_to_english(hungarian_text):
    result = translator.translate_text(hungarian_text, target_lang="en-US")
    return result.text

hungarian_text = open("./extracted_lyrics/lyrics.txt", "r", encoding="utf-8").read()
english_text = translate_to_english(hungarian_text)
print(english_text)
with open("./translated_lyrics/lyrics.txt", "w", encoding="utf-8") as file:
    file.write(english_text)
```
